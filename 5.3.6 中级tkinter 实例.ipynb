```python
#文件遍历系统
#！/usr/bin/env python

import os
from time import sleep
from tkinter import *

class DirList(object):
    
    def __init__(self,initdir=None):
        self.top = Tk()
        self.label = Label(self.top,text="Directory Lister v1.1") #标题
        self.label.pack()
        
        self.cwd = StringVar(self.top) #cwd 变量用于保存当前所在的目录名
        
        self.dir1 = Label(self.top,fg = "blue",font = ("Helvetica",12,"bold")) #用于显示当前目录名
        self.dir1.pack()
        
        self.dirfm = Frame(self.top)
        self.dirsb = Scrollbar(self.dirfm)
        self.dirsb.pack(side=RIGHT,fill=Y)#放在Frame的右边，填充Y方向
        self.dirs = Listbox(self.dirfm,height=15,width=50,yscrollcommand=self.dirsb.set)#Listbox控件，列出目录中的文件列表
        self.dirs.bind("<Double-1>",self.setDirAndGo)#与回调函数setDirAndGo连接起来，<Double-1>鼠标双击事件，执行bind的函数
        self.dirsb.config(command=self.dirs.yview)#作用于Listbox的Y轴
        self.dirs.pack(side=LEFT,fill=BOTH)
        self.dirfm.pack()
        
        self.dirn = Entry(self.top,width=50,textvariable=self.cwd) #创建一个文本输入框，用户可以输入想要遍历的目录名
        self.dirn.bind("<Return>",self.doLS) #绑定回车事件
        self.dirn.pack()
        
        self.bfm = Frame(self.top) #定义一个按钮的框架，放置三个按钮“clear”“go按钮”“quit按钮”
        self.clr= Button(self.bfm,text="Clear",
                        command = self.clrDir,fg="white",bg='blue') #配置clrDir回调函数
        self.ls = Button(self.bfm,text="List Directory",
                        command = self.doLS,fg = 'white',bg='green') #配置doLS回调函数
        self.quit = Button(self.bfm,text="Quit",command = self.top.quit, fg="white",bg = "red") #配置quit ，退出
        self.clr.pack(side=LEFT)
        self.ls.pack(side=LEFT)
        self.quit.pack(side=LEFT)
        self.bfm.pack()
        
        if initdir:
            self.cwd.set(os.curdir)
            self.doLS() #初始化，以当前目录作为GUI开始的起点
            
    def clrDir(self,ev=None):
        self.cwd.set("") #清空Tk字符串变量cwd，ev参数是由窗口系统传入的，默认是None
        
    def setDirAndGo(self,ev=None): #设置要遍历的目录，并通过调用doLS实现遍历目录的行为？
        self.last = self.cwd.get()
        self.dirs.config(selectbackground="red")
        check = self.dirs.get(self.dirs.curselection())
        if not check:
            check = os.curdir
        self.cwd.set(check)
        self.doLS()
        
    def doLS(self,ev=None):
        error = ""
        tdir = self.cwd.get()
        if not tdir:tdir = os.curdir
        if not os.path.exists(tdir):
            error = tdir + ":no such file"
        elif not os.path.isdir(tdir):
            error = tdir + ":not a directory"
            
        if error:
            self.cwd.set(error)
            self.top.update()
            sleep(2)
            if not (hasattr(self,"last") and self.last):
                self.last = os.curdir
            self.cwd.set(self.last)
            self.dirs.config(selectbackground="LightSkyBlue")
            self.top.update()
            return
        self.cwd.set("FETCHING DIRECTORY CONTENTS...")
        self.top.update()
        #调用os.listdir获取实际文件列表并在Listbox中进行替换   
        dirlist = os.listdir(tdir)
        dirlist.sort()
        os.chdir(tdir)
            
        self.dir1.config(text=os.getcwd())
        self.dirs.delete(0,END)
        self.dirs.insert(END,os.curdir)
        self.dirs.insert(END,os.pardir)
        for eachFile in dirlist:
            self.dirs.insert(END,eachFile)
        self.cwd.set(os.curdir)
        self.dirs.config(selectbackground="LightSkyBlue")
            
def main():
    d = DirList(os.curdir)
    mainloop()
    
if __name__ == "__main__":
    main()
```
